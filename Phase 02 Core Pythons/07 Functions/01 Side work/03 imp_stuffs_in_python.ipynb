```python
#  let's Rewrite the palindrome code again: and this concept is called the functional approach (non primitive approach)

def Palindrome_checker(a): # parameter
    copy = a
    rev = 0

    while a > 0:
        rev = rev * 10 + a % 10
        # a = a / 10  Don't do this you will have a lot of error's
        a = a // 10

    if copy == rev:
        print(f"{copy} is a Palindrome number")
    else:
        print(f"{copy} is No palindrome number ")


Palindrome_checker(121) #arguments
Palindrome_checker(333)
Palindrome_checker(632367454)
```

```python
#  parameters and arguments
#  Parameters are the values you accept while calling the function
#  arguments are the values you provide to parameters while calling the function
    
# Types of arguments:
# 1. Positional arguments  - order matters
# 2. Default - Works even without Passing values
# 3. Keyword - Pass in any order





```

```python
# 1. Positional keyword:
def multiply(a, b, c, d):
    print(a * b * c * d)


multiply(2, 3, 4, 10)  # Your output for this function call is : 240
multiply(10, 2, 10, 3)  # Your Output for this function call is : 600
multiply(7, 5, 3, 2)  # Your Output for this function call is : 210

# Note:  Error possible in the function call us : TypeError : function() missing x required positional argument : X,...
```

```python
# 2. Default - Works even without Passing values
def addition(a,b,c=12):
    print(a+b+c)



addition(5,6) # 23
addition(4,5,3) # 12 , and value of the c is going to overwrite , from c= 12 -> 3 
```
